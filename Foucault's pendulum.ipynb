# Маятник Фуко
### Уравнение маятника Фуко имеет следующий вид:
![image](image.png)

1. Сначала импортируем необходимые библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
```

2. Введем параметры для вычислений 

```python
N = 200 # кол-во шагов
# временной интервал
t_init = 0
t_end  = 180
# временной шаг 
dt = float(t_end - t_init) / N
# массив данных по времени
t = np.arange(t_init, t_end + dt, dt)   
L = 50 # длина нити 
Omega = 0.04 # циклическая частота колебний
lamda = 0.5*np.pi
g=9.8 #  ускорение свободного падения
```

3. Реализация вычисления уравнения маятника Фуко

```python
# возьмем произвольные значения 0.7 и 1 для с1 и с2 соответственно  
с1=0.7
с2=1
    
e1=np.exp(1j*(np.sqrt(g/L)*t)) # реализация exp(i*sqrt(g/L)*t)
e2=np.exp(1j*((-1)*np.sqrt(g/L)*t)) # реализация exp(-i*sqrt(g/L)*t) 
e3=np.exp(1j*(Omega*np.sin(lamda)*(-1)*t)) # реализация exp(-i*Omega*sin(lamda))

# решение уравнения 
p=(с1*e1+с2*e2)*e3
```

4. Траектория маятника Фуко

```python
# отделение действительной части чисел 
x=p.real
y=p.imag
    
plt.plot(x,y,)
plt.xlabel("X")
h = plt.ylabel("Y")
h.set_rotation(0)
plt.grid()
plt.show()
```
